```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Datos del vaso de cartón pequeño
data_vaso_c = list()
for i in range(5):
    data_vaso_c.append(pd.read_csv("Datos vasos/tabla_vaso_carton%s.txt" % str(i+1), delimiter = ";"))
```

```python
for i in range(5):
    plt.plot(data_vaso_c[i]["t"], data_vaso_c[i]["x"], label = "vaso %s" % str(i+1))
    plt.title("Vaso cartón pequeño, Distancia")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Posición (m)")
    
plt.legend()
```

```python
for i in range(5):
    plt.plot(data_vaso_c[i]["t"], data_vaso_c[i]["vx"], label = "vaso %s" % str(i+1))
    plt.title("Vaso cartón pequeño, Velocidad")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Velocidad (m/s)")
    
plt.legend()
```
